```python
#Helper function to fetch Director name from field "crew".
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L 
```

```python
#Helper function to remove spaces in between any names or entities so that it wont split during tokenization of words
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
df['cast'] = df['cast'].apply(collapse)
df['crew'] = df['crew'].apply(collapse)
df['genres'] = df['genres'].apply(collapse)
df['keywords'] = df['keywords'].apply(collapse)
```

```python
from gensim.models import Word2Vec
from sklearn.feature_extraction.text import CountVectorizer
```

### WORD2VEC

```python
# tags data
data = new_df['tags']

# Tokenize the text
tokenized_data = [text.split() for text in data]

# Train a Word2Vec model
word2vec_model = Word2Vec(sentences=tokenized_data, vector_size=100, window=5, min_count=1, workers=4)

# Function to get the average word vector for a document
def get_average_word_vector(words, model, num_features):
    feature_vector = np.zeros((num_features,), dtype="float32")
    num_words = 0
    index2word_set = set(model.wv.index_to_key)

    for word in words:
        if word in index2word_set:
            num_words += 1
            feature_vector = np.add(feature_vector, model.wv[word])

    if num_words != 0:
        feature_vector = np.divide(feature_vector, num_words)

    return feature_vector

# Get word vectors for each document
word_vectors = [get_average_word_vector(words, word2vec_model, 100) for words in tokenized_data]

# Convert word vectors to a numpy array
word_vectors_matrix = np.array(word_vectors)
```

### Calculating similarity matrix

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(word_vectors_matrix)
```

```python
similarity[0]
```

```python
#Helper function to recommend movie, by calculating similarity scores

def recommend_movie(movie):
    index = new_df[new_df['title'] == movie].index[0]                                         #Finding index of the input movie
    distances = sorted(list(enumerate(similarity[index])), reverse=True, key=lambda x: x[1])  #Calculating distances (similarity measure from that movie to each and every movie)
    print(f"These are the top 5 recommended movies based on your input '{movie}':")
    print("-" * 70)
    for i in distances[1:6]:                      #Looping only through next 5 distances in the array
        movie_title = new_df.iloc[i[0]].title     #Extracting the titles from new_df(dataframe)
        similarity_score = i[1]                   #Extracting similarity scores
        print(f"Movie: {movie_title}, Similarity Score: {similarity_score}")
```
